>How did you preprocess data, e.g. cleaning, feature engineering, etc?

導入train.csv以後首先將標籤{-1,1}映射為{0,1},並按照8:2劃分為train set 和 val set 

然後用 Beautifulsoup 洗去 HTML 標籤，接著用正則表達式刪除非單詞字符並同意轉化成小寫字母，最初嘗試了很多種特徵，經過反復實驗最後採用一下特徵： 'Channel', 'Topic', 'Day', 'Date', 'Month', 'Year', 'Hour', 'Content_Len', 'Num_Image'.

其中'Channel' 和 'Topic' 為文本特徵，用BOW向量化,'Day', 'Date', 'Month', 'Year', 'Hour', 'Content_Len', 'Num_Image'則是數字特徵可以直接填充。

提取完特徵後將左右特徵拼接成一個，然後準備進入模型訓練，

>How did you build the classifier, e.g. model, training algorithm, special techniques, etc?

採用了random forest,lightGBM,XGBoost和CatBoost四個模型加權投票的處理，所有模型共享資料的預處理，確保訓練後結果的AUC具有參考意義。

在每個模型訓練完成後輸出trian AUC 和 val AUC,根據實際表現調整各個模型投票的權重,經過反復實驗發現以lightGBM為主要模型，其他模型只給予輕權重0.1左右，此時模型表現較好。

>Conclusions, including interesting findings, pitfalls, takeaway lessons, etc.

一開始我只是簡單按照網頁的指導，將資料導入後洗去 HTML 標籤，直接將title channel 等部分分別用特徵哈希轉換成對應向量並拼接，然後放進模型中訓練，結果訓練結果很差，即使反復微調參數也無濟於事。

於是我將大部分內容特徵直接剔除，例如content author這類很難簡單提取特徵的模型，但是提升效果還是不明顯。

後來我去網上查找預測新聞流行度相關的論文和資料，發現大部分模型都只是抓取一些很簡單的特徵，尤其是與時間先關的特徵(因為新聞的時效性很強)和元素是否多樣化(是否有圖片、超鏈接),參考了多方信息後我重新設計了特徵，並採用多模型投票，最後取得了較好的效果。

實踐證明，良好的特徵設計以及資料預處理，遠比模型選擇和參數重要的多 : )
